```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Gender": ["Male", "Female"],
    "Count": [55, 45]
}

df = pd.DataFrame(data)

plt.figure(figsize=(6,4))
plt.bar(df["Gender"], df["Count"])
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```
